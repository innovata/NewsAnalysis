```python
def 위치정보에_대한_추가용어번역(df):
    # 용어번역 : 값이 아닌 문자열을 찾아서 번역

    add_dic = {
        '네이버_모바일뉴스_홈':'Naver_mNews_Home',
        '많이본뉴스':'popular_news',
        '썸네일뉴스':'thumbnail_news',
        '네이버_모바일홈':'Naver_mHome',
        '문자형_뉴스리스트':'character-type_news_list',
    }

    keys = list(add_dic.keys())
    for key in keys:
        print(add_dic[key])
        df['location'] = df['location'].str.replace(pat=key, repl=add_dic[key], n=-1, case=None, flags=0, regex=True)
        
    return df
```

### histogram_features.ipynb

```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

# example data
#mu = 100  # mean of distribution
mu = mean
#sigma = 15  # standard deviation of distribution
sigma = std

# x : list
#x = mu + sigma * np.random.randn(437)
x = list(df11_2['위치점유시간'])
print('\n x_type :\n', type(x))
print('\n x_len :\n', len(x))
#print('\n x :\n', x)

#num_bins = 50
num_bins = int(len(x)/10)

fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=1)

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, y, '--')
#ax.set_xlabel('Smarts')
ax.set_xlabel('occupation_time_of_news')
ax.set_ylabel('density')
ax.set_title(r'Histogram of IQ: $\mu={}s$, $\sigma={}$s'.format(int(mean), int(std)))

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```
