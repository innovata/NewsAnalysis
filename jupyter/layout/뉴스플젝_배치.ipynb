```python
# 오픈 패키지
import threading
import time
import schedule
import sys

# 나의 패키지
# 프로젝트 라이브러리
sys.path.append('/Users/sambong/p/news/')
from collect_naver_m_home import 네이버_모바일홈_뉴스영역을_수집후_저장
from collect_naver_m_newsHome import 네이버_모바일_뉴스홈을_수집후_저장
# 전역변수

eve_sec = 60
eve_mins = 10
```

```python
def run_threaded(job_func):
    job_thread = threading.Thread(target=job_func)
    job_thread.start()


schedule.every(eve_mins).seconds.do(run_threaded, 네이버_모바일홈_뉴스영역을_수집후_저장)
schedule.every(eve_mins).seconds.do(run_threaded, 네이버_모바일_뉴스홈을_수집후_저장)

while 1:
    schedule.run_pending()
    time.sleep(eve_mins)
```
