```python
def pd_read_html(url_dic):
    # 쓸모없는 함수 
    url_li = list(url_dic.values())
    for url in (url_li):
        print('='*100)
        print(url)
        try:
            df = pd.read_html(str(url))
            print(df)
        except Exception as e:
            print(e)
        
    
pd_read_html(url_dic)
```
