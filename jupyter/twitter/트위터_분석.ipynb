```python
def 로딩_home_line(fname):
    tweet_li = []
    with open(data_path+fname, 'r') as f:
        for line in f:
            tweet = json.loads(line)
            tweet_li.append(tweet)
    return tweet_li
```

```python
def 파싱_timeline(df):
    # https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
    df.created_at = df.created_at.apply(lambda x: datetime.strptime(x, '%a %b %d %H:%M:%S %z %Y'))
    return df
```

```python
ht_dicli = 로딩_home_line('home_timeline.jsonl')
ut_dicli = 로딩_home_line('user_timeline_cheolsoo0919.jsonl')
```

```python
df1 = 파싱_timeline(pd.DataFrame(ht_dicli)).sort_values(by='created_at', ascending=False)
df2 = 파싱_timeline(pd.DataFrame(ut_dicli)).sort_values(by='created_at', ascending=False)
```
