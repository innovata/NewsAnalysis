```python
# 수집-파싱-저장 한번에 가기.

def 수집(d):
    print('\n'+'= '*30+'수집')
    FormData = {
        's_poly_cd': d['정당코드'],
        's_dept_cd': None,
        's_dtl_no': None,
        's_elected_method': None,
        's_up_orig_cd': None,
        's_dw_orig_cd': None,
        's_mem_nm': None,
        'currentPage': 1,
        'movePageNum': None,
        'rowPerPage': 300,
    }
    print(FormData)
    r = requests.post(URL1, data=FormData)
    return r

def 국회의원현황_수집파싱저장():
    dicli = mg.find(db명=DB명, tbl명='정당명코드', query=None, projection=None, dbg_on=False, 컬럼순서li=[], df보고형태='dicli')
    dicli_len = len(dicli)
    i=1
    for d in dicli:
        print('\n'+'-'*60+'{}/{}, 정당명:{}'.format(i, dicli_len, d['정당명']))
        print(d)
        r = 수집(d)
        
        print('\n'+'= '*30+'파싱')
        soup = BeautifulSoup(r.text, 'html.parser')
        s = soup.find(class_='memberna_list')
        dl_tag_li = s.find_all('dl')
        dl_tag_li_len = len(dl_tag_li)
        j=1
        saving_dicli = []
        for dl in dl_tag_li:
            print('- '*30+'{}/{}'.format(j, dl_tag_li_len))
        #    print(dl.attrs)

            name_li = []
            dt = dl.find('dt')
            for string in dt.stripped_strings:
    #            print(string)
                name_li.append(string)

            etcInfo_li = []
            dd_li = dl.find_all('dd')
            for dd in dd_li:
                for string in dd.stripped_strings:
    #                print(string)
                    etcInfo_li.append(string)

        #    print(name_li)
        #    print(etcInfo_li)
            dic = {
                '정당명':d['정당명'],
                '이름':name_li[0],
                '이름CH':name_li[1],
                '이름EN':name_li[2],
                '초재선':etcInfo_li[0],
                '지역구':etcInfo_li[1],
            }
            print(dic)
            saving_dicli.append(dic)
    #        print('- '*30)
    #        print(dl.prettify())
            j+=1
    #        break

        print('\n'+'= '*30+'저장')
        df = pd.DataFrame(saving_dicli)
        mg.insert_many(db명=DB명, tbl명='국회의원현황', dicli=saving_dicli, dbg_on=False, 사전검증=False)
        i+=1
#        break
```
